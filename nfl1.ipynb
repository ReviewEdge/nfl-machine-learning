```python
import pandas as pd

df = pd.read_csv("nflData.csv", sep=",")
# shows top 10 rows
df.head(10)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python



home_team_winning = 0
away_team_winning = 0
tie_game = 0


def print_out():
    print(f"Home: {home_team_winning}\tAway: {away_team_winning}\t Tie: {tie_game}")


for r in range(len(df)):
    if(r%10000 == 0):
        print(f"...{r}...")
        print_out()
    play = df.iloc[r]
    # print(f"\nPlay ID: {play['play_id']}")
    if play["total_home_score"] > play["total_away_score"]:
        # print("Home Team Winning")
        home_team_winning+=1
    elif play["total_home_score"] < play["total_away_score"]:
        # print("Home Team Winning")
        away_team_winning+=1
    else:
        # print("Tie Game")
        tie_game+=1


print("\n\n\n")
print_out()

totals = [home_team_winning, away_team_winning, tie_game]

```

```python
print(totals)
```

```python
winning = ["Home Team Winning", "Away Team Winning", "Tie Game"]
colors = ["blue", "red", "gray"]
plt.bar(winning, totals, color=colors)
plt.title('# of Plays Winning for Home Team vs. Away Team')
plt.xlabel('Winning')
plt.ylabel('Total Plays')
plt.show()
```

```python
y = np.array(totals)

sum_plays = totals[0]+totals[1]+totals[2]

perc = []
for t in totals:
    p = t / sum_plays
    p*=100
    perc.append(p)
    
# print(perc)
    

pie_lab = [f"Home Team Winning - {round(perc[0], 1)}%\n({home_team_winning})", f"Away Team Winning - {round(perc[1], 1)}%\n({away_team_winning})", f"Tie Game - {round(perc[2], 1)}%\n({tie_game})"]

plt.pie(y, labels = pie_lab, colors=colors)
plt.show()
```

```python

passing = 0
win_and_pass = 0
lose_and_pass = 0
tie_and_pass = 0

run = 0
win_and_run = 0
lose_and_run = 0
tie_and_run = 0



for r in range(len(df)):
    if(r%10000 == 0):
        print(f"...{r}...")
        # print_out()
        
    play = df.iloc[r]
    
    pt=play["posteam_type"]
    play_type = play["play_type"]
    if play_type == "pass":
        passing+=1
    elif play_type == "run":
        run+=1
    
    winning_side = "tie"
        
    if play["total_home_score"] > play["total_away_score"]:
        winning_side = "home"
    elif play["total_home_score"] < play["total_away_score"]:
        winning_side = "away"
    
    # game is tied
    if winning_side == "tie":
        if play_type == "pass":
            tie_and_pass+=1
        elif play_type == "run":
            tie_and_run+=1
    
    # post team is winning
    elif pt==winning_side:
            if play_type == "pass":
                win_and_pass+=1
            elif play_type == "run":
                win_and_run+=1
    # post team is losing
    else:
        if play_type == "pass":
            lose_and_pass+=1
        elif play_type == "run":
            lose_and_run+=1



run_totals = [
    run,
    win_and_run,
    lose_and_run,
    tie_and_run
]

pass_totals = [
    passing,
    win_and_pass,
    lose_and_pass,
    tie_and_pass
]
```

```python
win_run_or_pass = win_and_pass + win_and_run
lose_run_or_pass = lose_and_pass + lose_and_run
tie_and_pass_or_run = tie_and_pass + tie_and_run
```
